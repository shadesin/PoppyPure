```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from skimage.feature import local_binary_pattern
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.decomposition import PCA

import joblib
```

```python
# 1. Dataset Path
# ----------------------------
DATASET_DIR = 'C:/Users/User/Desktop/ML Final year/Dataset'  # update if needed
```

```python
# ----------------------------
# 2. Feature Extraction
# ----------------------------
def extract_features(image, bins=(8, 8, 8)):
    """
    Extracts color histogram (HSV) and texture features (LBP) from the given image.
    """
    image = cv2.resize(image, (128, 128)) # The input image is resized to a fixed dimension of 128x128 pixels

    # Color: HSV histogram
    hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) # The resized image is converted from its original BGR (Blue, Green, Red) color space
    hist = cv2.calcHist([hsv_img], [0, 1, 2], None, bins, [0, 180, 0, 256, 0, 256]) # This is the core of HSV histogram calculation.
    hist = cv2.normalize(hist, hist).flatten() # cv2.normalize(hist, hist): The calculated histogram is normalized

    # Texture: Local Binary Pattern
    gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # For texture analysis, the image is first converted to grayscale.
    lbp = local_binary_pattern(gray_img, P=8, R=1, method='uniform') #This line calculates the Local Binary Pattern. 
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 10), range=(0, 9)) # A histogram of the LBP image is calculated.
    lbp_hist = lbp_hist.astype("float") # The LBP histogram values are cast to floating-point numbers. 
    lbp_hist /= (lbp_hist.sum() + 1e-6) # The LBP histogram is normalized by dividing each bin count by the total sum of all bin counts.

    return np.concatenate([hist, lbp_hist])

```

```python
# After extracting features:
X = []
y = []

for folder in os.listdir(DATASET_DIR):
    folder_path = os.path.join(DATASET_DIR, folder)
    if os.path.isdir(folder_path):
        for filename in os.listdir(folder_path):
            img_path = os.path.join(folder_path, filename)
            image = cv2.imread(img_path)
            if image is None:
                print("Image not loaded:", img_path)
                continue
            features = extract_features(image)
            if len(features) == 0:
                print("No features extracted for:", img_path)
                continue
            X.append(features)
            y.append(folder)

X = np.array(X)
y = np.array(y)

print("Final Feature Shape:", X.shape)
print("Final Labels Shape:", y.shape)
```

```python
# Apply PCA to retain 95% variance
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X)  # X is your original feature set (HSV + LBP)

# Now split the PCA-transformed data
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = { #This dictionary defines the hyperparameter search space. GridSearchCV will try every possible combination of these parameters.
    'hidden_layer_sizes': [(256, 128), (512, 256), (128, 64)], #Defines the architecture of the hidden layers in the MLP. Each tuple represents the number of neurons in each hidden layer.
    'activation': ['relu', 'tanh'],
    'alpha': [0.0001, 0.001, 0.01], #L2 regularization term. It adds a penalty to the loss function that shrinks the model coefficients towards zero, which helps prevent overfitting.
    'solver': ['adam', 'sgd'], #The algorithm for weight optimization. adam is an adaptive learning rate optimizer, generally robust. 
    'learning_rate_init': [0.001, 0.01] #controls the step size at each iteration while moving towards a minimum of the loss function.
}

grid = GridSearchCV(MLPClassifier(max_iter=500, random_state=42),
                    param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best CV Accuracy:", grid.best_score_)

# Use best estimator
mlp = grid.best_estimator_
mlp.fit(X_train, y_train)
```

```python
# ----------------------------
# 4. Encode Labels
# ----------------------------
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
# ----------------------------
# 5. PCA Visualization
# ----------------------------
print("Reducing dimensions using PCA for visualization...")
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='tab10', edgecolors='k', alpha=0.7)
plt.legend(*scatter.legend_elements(), title="Class")
plt.title("PCA Projection of Feature Space")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ----------------------------
# 6. Split into Train/Test
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)

```

```python
# ----------------------------
# 7. Train MLP Classifier
# ----------------------------
print("Training MLP Classifier...")
mlp = MLPClassifier(
    hidden_layer_sizes=(256, 128),
    activation='relu',
    solver='adam',
    max_iter=500,
    random_state=42
)
mlp.fit(X_train, y_train)
```

```python
# ----------------------------
# 8. Evaluation
# ----------------------------
print("\nModel Evaluation:")
y_pred = mlp.predict(X_test)

print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
# ----------------------------
# 9. Confusion Matrix
# ----------------------------
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

```python
# 10. Save Model
# ----------------------------
joblib.dump(mlp, "mlp_color_texture.pkl")
joblib.dump(le, "label_encoder3.pkl")
```
